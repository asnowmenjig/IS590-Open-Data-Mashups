```python
# EEBO

import pandas
import numpy

eebo_df = pandas.read_csv('../eebo/DO-NOT-EDIT/eebo_dataset_final_DO-NOT-EDIT.csv', nrows=40)
# eebo_df
```

```python
eebo_df2 = eebo_df.loc[:, ['first_name', 'last_name', 'fraction_total']]
# eebo_df2
```

```python
eebo_df3 = eebo_df2.assign(dataset='eebo')
eebo_df3
```
